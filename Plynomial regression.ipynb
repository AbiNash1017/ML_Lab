```python
# Importing libraries
import matplotlib.pyplot as mtp
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Importing dataset
data_set = pd.read_csv('Dataset/kc_house_data.csv')

# Assuming 'price' is the target variable and 'sqft_lot' is the feature
# Define x and y based on the dataset
x = data_set[['sqft_lot']].values  # Feature matrix
y = data_set['price'].values       # Target variable

# Fitting Polynomial Regression
poly_regs1 = PolynomialFeatures(degree=2)
x_poly = poly_regs1.fit_transform(x)

lin_regs_3 = LinearRegression()
lin_regs_3.fit(x_poly, y)

# Plotting the Polynomial Regression results
mtp.scatter(x, y, color='green')
mtp.plot(x, lin_regs_3.predict(x_poly), color='pink')
mtp.title("Price Prediction Model (Polynomial Regression)")
mtp.xlabel("sqft_lot")
mtp.ylabel("price")
mtp.show()

# Prediction for a new level
# Ensure the input for prediction has the correct shape
lin_pred1 = lin_regs_3.predict(poly_regs1.transform([[11]]))
print("Prediction value for new level using polynomial regression:", lin_pred1)
```

```python
#POLYNOMIAL REGRESSION
import numpy as nm
import pandas as pd
import matplotlib.pyplot as mtp
from sklearn.linear_model import LinearRegression
data_set = pd.read_csv('Dataset/Position_Salaries.csv')
print(data_set.head())
x = data_set.iloc[:, 1:2].values
y = data_set.iloc[:, 2].values
print(x, y)
from sklearn.preprocessing import PolynomialFeatures
ply_regs = PolynomialFeatures(degree=2)
x_poly = ply_regs.fit_transform(x)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly, y)
mtp.scatter(x, y, color="blue")
mtp.plot(x, lin_reg_2.predict(x_poly), color="red")
mtp.title("Bluff Detection Model (Polynomial Regression)")
mtp.xlabel("Position Levels")
mtp.ylabel("Salary")
mtp.show()
lin_pred = lin_reg_2.predict(ply_regs.transform([[11]]))
print("Prediction value for new level using polynomial regression:", lin_pred)
```
